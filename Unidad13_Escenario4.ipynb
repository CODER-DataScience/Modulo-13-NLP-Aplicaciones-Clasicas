# Escenario 4: Automatización de soporte (Chatbots)

## 1. Perspectiva de negocio
El equipo de soporte recibe muchas consultas repetitivas (horarios, precios, políticas).  
El reto es **automatizar respuestas frecuentes** para:
- Reducir la carga de los agentes humanos.
- Brindar atención inmediata 24/7.
- Mejorar la satisfacción del cliente.

## 2. Interpretación esperada
- Si el cliente pregunta por "horario" → el chatbot responde con el horario de atención.  
- Si pregunta por "precio" → devuelve la información de precios.  
- Si pregunta por "contacto" → ofrece datos de contacto.  
- Si no reconoce la consulta → responde con un mensaje genérico.

## 3. Valor de negocio
- **Eficiencia**: libera tiempo de los agentes para casos complejos.  
- **Disponibilidad**: atención inmediata en cualquier momento.  
- **Satisfacción**: reduce tiempos de espera y mejora la experiencia del cliente.  


```python
# Chatbot simple basado en reglas

def chatbot(pregunta):
    pregunta = pregunta.lower()
    if "horario" in pregunta:
        return "Nuestro horario de atención es de lunes a viernes de 9 a 18 hs."
    elif "precio" in pregunta:
        return "El precio del servicio básico es de $1000 mensuales."
    elif "contacto" in pregunta:
        return "Puedes contactarnos al correo soporte@empresa.com."
    else:
        return "Lo siento, no tengo una respuesta para esa consulta. ¿Podrías reformularla?"

# Ejemplos de prueba
preguntas = [
    "¿Cuál es el horario de atención?",
    "Necesito saber el precio del plan",
    "¿Cómo puedo contactar al soporte?",
    "¿Tienen promociones especiales?"
]

for p in preguntas:
    print(f"Cliente: {p}")
    print(f"Chatbot: {chatbot(p)}\n")
```

## Conclusiones del resultado

### 1. Respuestas correctas
- "¿Cuál es el horario de atención?" → ✅ El chatbot respondió con el horario correcto.  
- "Necesito saber el precio del plan" → ✅ El chatbot devolvió el precio esperado.  

👉 El sistema funciona bien para consultas que coinciden con las palabras clave definidas.

### 2. Respuestas incorrectas / limitaciones
- "¿Cómo puedo contactar al soporte?" → ❌ El chatbot no reconoció la palabra "contactar" y devolvió un mensaje genérico.  
- "¿Tienen promociones especiales?" → ❌ No hay reglas para esta consulta, por lo que también devolvió un mensaje genérico.  

👉 El chatbot basado en reglas simples **no entiende variaciones de lenguaje** ni consultas fuera de su lista de palabras clave.

### 3. Insight general
- Este enfoque es útil como **primer prototipo** para automatizar respuestas frecuentes.  
- Sin embargo, es limitado: depende de coincidencias exactas y no maneja sinónimos ni contexto.  
- Para mejorar, se puede:
  - Ampliar el diccionario de palabras clave.  
  - Usar NLP más avanzado (intents + entidades).  
  - Entrenar un modelo con ejemplos reales de consultas de clientes.

### 4. Valor de negocio
- **Eficiencia**: reduce la carga de los agentes humanos en preguntas repetitivas.  
- **Disponibilidad**: atención inmediata 24/7.  
- **Satisfacción**: mejora la experiencia del cliente al dar respuestas rápidas.  
- **Escalabilidad**: se puede integrar en múltiples canales (web, WhatsApp, apps).
