# Escenario 2: Organización de noticias (Clasificación de texto)

## 1. Perspectiva de negocio
Un portal de noticias recibe gran cantidad de artículos diariamente.  
El reto es **clasificarlos automáticamente por temática** para:
- Mejorar la experiencia del lector.
- Facilitar búsquedas y organización interna.
- Ahorrar tiempo antes de aplicar modelos más avanzados.

## 2. Interpretación esperada
- Textos con palabras como "dólar", "inflación", "bonos" → economía.  
- Textos con "equipo", "partido", "campeonato", "delantero" → deportes.  
- Otros temas → categoría "otros".

## 3. Valor de negocio
- Permite organizar grandes volúmenes de noticias sin intervención manual.  
- Mejora la experiencia del usuario al encontrar rápidamente lo que busca.  
- Es un primer paso antes de usar modelos más sofisticados (Naive Bayes, Transformers).


```python
# Clasificador simple por reglas

def clasificar(texto):
    texto = texto.lower()
    if any(p in texto for p in ["dólar", "inflación", "bonos", "consumo"]):
        return "economía"
    elif any(p in texto for p in ["equipo", "partido", "campeonato", "delantero"]):
        return "deportes"
    else:
        return "otros"

# Ejemplos de prueba
ejemplos = [
    "El partido fue intenso y muy disputado",
    "La economía muestra signos de recuperación",
    "El delantero marcó dos goles",
    "Se inauguró una nueva exposición de arte",
    "La inflación sigue afectando el consumo",
    "El equipo logró una victoria histórica"
]

for texto in ejemplos:
    print(f"Texto: {texto}")
    print(f"Clasificación: {clasificar(texto)}\n")
```

## Conclusiones del resultado

### 1. Clasificaciones correctas
- "El partido fue intenso y muy disputado" → **deportes** ✅  
- "El delantero marcó dos goles" → **deportes** ✅  
- "La inflación sigue afectando el consumo" → **economía** ✅  
- "El equipo logró una victoria histórica" → **deportes** ✅  

👉 El clasificador por reglas simples funciona bien cuando las palabras clave están presentes.

### 2. Clasificaciones incorrectas / limitaciones
- "La economía muestra signos de recuperación" → **otros** ❌  
- "Se inauguró una nueva exposición de arte" → **otros** (correcto, pero muestra que el modelo no cubre más categorías).  

👉 El clasificador falla cuando el texto no contiene exactamente las palabras clave definidas (ej. “economía” no estaba en la lista de palabras).

### 3. Insight general
- Este enfoque **no entiende el contexto**, solo busca coincidencias de palabras.  
- Es útil como **primer paso rápido** para organizar grandes volúmenes de texto.  
- Para mejorar, se necesitaría ampliar el diccionario de palabras clave o usar modelos más avanzados (Naive Bayes, embeddings, Transformers).

## Valor para el negocio
- **Ahorro de tiempo**: permite organizar noticias automáticamente sin intervención manual.  
- **Escalabilidad**: se puede aplicar a miles de artículos diarios.  
- **Base inicial**: aunque limitado, sirve como punto de partida antes de entrenar clasificadores más sofisticados.
